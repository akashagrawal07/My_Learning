```python
import pandas as pd
import numpy as np

from pandas import Series,DataFrame
```

# Finding what data is missing

```python
missing = np.NaN
series_obj = Series(['row1','row2',missing,'row4','row5',missing,'row7','row8'])
series_obj
```

```python
series_obj.isnull()
```

# Filling Missing Value

```python
np.random.seed(25)
df_obj = DataFrame(np.random.rand(36).reshape((6,6)))
df_obj
```

```python
df_obj.loc[3:5,0] = missing
df_obj.loc[1:3,5] = missing
df_obj
```

```python
fil_df = df_obj.fillna(0)
fil_df
```

```python
fil_df = df_obj.fillna({0:7,5:1.5})
fil_df
```

# Counting Missing Value

```python
df_obj.isnull().sum()
```

# Filtering out Missing values

```python
drop_df_row = df_obj.dropna() 
drop_df_row # it will drop only row which don't have NULL value
```

```python
drop_df_col= df_obj.dropna(axis=1) 
drop_df_col # it will drop only col which don't have NULL value
```
