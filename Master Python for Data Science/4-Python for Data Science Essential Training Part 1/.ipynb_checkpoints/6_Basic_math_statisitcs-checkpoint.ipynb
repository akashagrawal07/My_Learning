# 3. Categorical Variable

Best Ways to Describe Categorical Variable
1. counts
2. Variable Description
3. Grouping

# The Basics

```python
address = 'mtcars.csv'

cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
cars.index = cars.car_names
cars.head(10000)
```

```python
carb = cars.carb
carb.value_counts()
```

```python
cars_cat = cars[['cyl','vs','am','gear','carb']]
# cars_cat = cars[['gear','carb']]
cars_cat.head()
```

```python
gear_group = cars_cat.groupby('gear')
# gear_group.groups()
gear_group.describe()
```

# Transforming Variables to categorical data type

```python
cars['group'] = pd.Series(cars.gear,dtype='category')
print(type(cars['group']))
cars['group'].dtype
```

```python
cars['group'].value_counts()
```

# Describing Categorical data with cross tabs

```python
pd.crosstab(cars['am'],cars['gear'])
# pd.crosstab(cars['am'],[cars['gear'],cars['carb']])
```
