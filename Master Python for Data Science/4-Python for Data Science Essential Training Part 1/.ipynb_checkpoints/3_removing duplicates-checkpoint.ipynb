# Removing Duplates

```python
df_obj = DataFrame({'Column 1':[1,1,2,2,2,3,4,5,5],
                    'Column 2':['a','a','b','b','b','c','d','e','e'],
                    'Column 3':['A','A','B','B','B','C','D','E','E'],
                   })
df_obj
```

```python
df_obj.duplicated()
```

```python
df_obj.drop_duplicates()
```
