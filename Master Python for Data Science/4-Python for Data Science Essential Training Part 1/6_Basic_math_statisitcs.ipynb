### Pearson correlation

Model means

 PC means Linear related ,Normaly distributed , continous numeric value 
    
    
Need to check - Bionomial variable , multinomial variable

Correlation is a statistical measure that expresses the extent to which two variables are linearly related (meaning they change together at a constant rate). It's a common tool for describing simple relationships without making a statement about cause and effect.


The Pearson coefficient is a measure of the strength and direction of the linear association between two variables with no assumption of causality. The Pearson coefficient shows correlation, not causation.    
    

### Calculated the pearson correlation sufficient

```python
mpg = cars['mpg']
hp = cars['hp']
qsec = cars['qsec']
wt = cars['wt']

pearsonr_coefficient,p_value = pearsonr(mpg,hp)
print('Pearson correlation Coefficient %0.3f'%pearsonr_coefficient)
```

#  Non-parametric correlation analysis

This analysis to find correlation between categorical , nonlinearly related , non-normally distributed variable

### Spearman's Rank Correlation

- Finds R Correlation between variable-pair of ordinal data type , 
   ordinal type - numeric variable that able to catergerise

- Variable pairs are then able to be ranked according to the strength of the correlation between them

R = 1 -> Strong positive relationship

R = 0 -> not linearly correlated

R = -1 -> Strong negative relationship


your variable are ordinal  numeric but able to be ranked like categorical variable

your variable are related non linearly

your data is non normally distributed

### Chi Square Test for independence

- p < 0.05 --> Reject null hypothesis and conclude  that the variable are correlated 

- p > 0.005 --> Accept null hypothesis and conclude that the variable are independent 

### User Chi Square Table Varaible if variable are categorical or numeric  and numeric variable are binned

 - Binning meaning created group suppose we have people from 10 - 90  we can binn them into 10-25,25-60,60-90 (like cowid vaccination age group is binning)

https://docs.tibco.com/pub/spotfire/7.0.1/doc/html/bin/bin_what_is_binning.htm


```python
import pandas as pd
import numpy as np
from numpy.random import randn
import random
from pandas import Series,DataFrame
import seaborn as sb
import scipy
from scipy.stats.stats import pearsonr
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 8,4

plt.style.use('seaborn-whitegrid')
```

```python
address = 'mtcars.csv'
cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
cars.head()
```

```python
sb.pairplot(cars)
```

```python
x = cars[['cyl','vs','gear']]
sb.pairplot(x)
```

```python
cyl = cars['cyl']
vs = cars['vs']
am = cars['am']
gear = cars['gear']

spearman_coefficient,p_value  = scipy.stats.spearmanr(cyl,vs)
print("Spearman correlation Coefficient %0.3f"%spearman_coefficient)
```

```python
spearman_coefficient,p_value  = scipy.stats.spearmanr(cyl,am)
print("Spearman correlation Coefficient %0.3f"%spearman_coefficient)
```

```python
spearman_coefficient,p_value  = scipy.stats.spearmanr(cyl,gear)
print("Spearman correlation Coefficient %0.3f"%spearman_coefficient)
```

### Chi Square test for independence

```python
table = pd.crosstab(cyl,am)
from scipy.stats import chi2_contingency

chi2,p,dof,excepted = chi2_contingency(table.values)
print("Chi Square Statistic %0.3f p_value %0.3f"%(chi2,p))
```

```python
table = pd.crosstab(cyl,vs)
from scipy.stats import chi2_contingency

chi2,p,dof,excepted = chi2_contingency(table.values)
print("Chi Square Statistic %0.3f p_value %0.3f"%(chi2,p))
```

#  Scaling and Transforming

### Scale your data

- So that differing magnitude among variable do not produce erroneous or misleading statistics
- To prepare data for machine learning

### Two ways to scale your data

1. Normalization: putting each observation on a relative scale between the value of 0 and 1

    (value of observation)/(sum of all observation in variable)
    
    
2. Standarizaton : Rescaling data so that it has zero mean and unit variance


### Using Scikit learn Preprocessing For

- Scaling your data
- Centering your data - centering a variable is subtracting the mean of the variable from each data point
- Normalizing your data - consider your marks on basis (6 question correct , 4 wrong) i'll get 6 marks out of 10
- Binning your data - 
- Imputing your data - a technique used for replacing the missing data with some substitute value

# Outlier Detection

- Most Machine learning methods asume that your has been treated for outlier
- Detecting outlne can be a data preprocessing task or an analytical method of its own merit

### Outlier Analysis Use cases
- Equipement failure
- Fraud
- CyberSecurity Event

### Univariate Method - Tukey Boxplot

- Tukey Boxplot (like candle chart with open , close , high and low)
    Boxplot whiskers are set at 1.5 (IQR)
        . upperquartile (like open) - 25% of data point are greater that this value
        . lower quartile(like close) - 25% of data point are less that this value
        . interquartile range IQR - middle of upper and lower quartile
        . in box plot Outlier  are data point which above high and lower than low 

. Inter quartile range (IQR)

- IQR - the distane between the first quratile(at 25%) and third quratile (at 75%)
  
### Univariate Method - Tukey Outlier Labeling

a = Q1 -1.5(IQR)

b = Q3 + 1.5(IQR)

if...
. min.value < a,or
. max.value > b then 
The variable is suspect for outlier
