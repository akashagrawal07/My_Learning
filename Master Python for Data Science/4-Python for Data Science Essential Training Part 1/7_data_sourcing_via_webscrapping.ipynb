# Working with ParasedData in Beautiful Soup

### Searching and Retrieving Data

- using find_all

```python
from bs4 import BeautifulSoup

import urllib
import urllib.request
import re
```

```python
with urllib.request.urlopen("https://github.com/ms-robot-please/Data-Mania-Demos/blob/master/IoT-2018.html") as response:
    html = response.read()

```

```python
soup = BeautifulSoup(html,'lxml')
type(soup)
```

### Parsing yor Data

```python
print(soup.prettify()[0:100])
```

### Getting data from parse tree

```python
text_only = soup.get_text()
print(text_only)
```

### Searching and retriving data from a parse tree

###### Retriving tag by filitering with name argument

```python
soup.find_all('li')
```

###### Retriving tag by filitering with keyword argument

```python
a = soup.find_all(href="https://stars.github.com")
a
```

###### Retriving tag by filitering with string argument

```python
soup = BeautifulSoup(html,'lxml')

soup.find_all("<ol>")
```

###### Retriving tag by filitering with list object

```python
soup.find_all(['ol','b'])
```

```python
t = re.compile('t')
for tag in soup.find_all(t):
    print(tag.name)
```

###### Retriving tag by filitering with boolean value

```python
t = re.compile('t')
for tag in soup.find_all(True):
    print(tag.name)
```

###### Retriving tag by filitering with string object

```python
for link in soup.find_all('a'):
    print(link.get('href'))
```

###### Retriving tag by filitering with regualr expression

```python
soup.find_all(string=re.compile('data'))
```

### Introduction to Natural Language Processing (NLTK)

Sentence Tokenization

    . it is process of breaking down paragraphs or a complete set of text into sentence
    
Word Tokenization

    . It is the process of breaking down a paragraph , a sentence or a complete text corupus into an array of words
    
Stop Words

    . Stop words are extremely common words that add little to little meaning to a body of text at a large sentence 
      stop words can be thought of as noise in the corpus of a body of text
      Example of Stop words include "a", these,they'd,do ,have ,you
 Stemming 
 
     . Stemming is a process for reducing the size of the corpus by converting words to their root word
       it is a basically a process for linguistic normalization.
       
       Raw text : run,ran, running (all words have meaning )
       Root word : run (we will use only root this is called stemming)
       
 Stems
 
     . this are simple parts of words
     
     
 Root Word
 
     . a stem word that may or may not have meaning when taken alone
     . root wird from which other words are derived
     . the primary form of word , upon which derivates are made
     
 Base Word
 
     . a base word that has meaning even when taken alone
     . it a base world that is used outright within a language, with only prefixes or suffixes being attached
     .simplest form of a word
     
 Lemmatizing
     
     . it is process of reducing a word to its base word.
     Stemming:
         Raw text : "Better"
         Root word : "Better"
     Lemmatizing :
         Raw text : "Better"
         Root word : "good" --> this is lemmatizing
         
Part of Speech tagging

    . process of identifying a part of speech within a gven body of text

Frequecy Distribution plotting

    . to understand of disctribution of words , used for sentiment analysis

```python
from nltk.tokenize import word_tokenize
word_tk = word_tokenize(text)
print("Word Tokenizing text - \n")
print(word_tk)
print(type(word_tk))
```

### Remove the Stop Words

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords

sw = set(stopwords.words('english'))
print("This are our stopwords:\n")
print(sw)
```

```python
filtered_stopwords = [w for w in word_tk if not w in sw]
print(filtered_stopwords)
```

### Stemming

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize,word_tokenize

portstem = PorterStemmer()
```

```python
stemmed_words = []

for w in filtered_stopwords:
    stemmed_words.append(portstem.stem(w))
    
print("Filtered Word",filtered_stopwords)
print()
print("Stemmed words",stemmed_words)
```
