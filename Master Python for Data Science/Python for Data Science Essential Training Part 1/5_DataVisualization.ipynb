```python
address = 'D:\Trading\Python_Algo_Trading\GIT\Master Python for Data Science\Python for Data Science Essential Training Part 1\Ex_Files_Python_Data_Science_EssT_Pt_1\Exercise Files\Data\mtcars.csv'
cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']

mpg = cars.mpg


fig = plt.figure()
ax = fig.add_axes([1,1,3,1])

mpg.plot(marker='*',mew=9)
ax.set_xticks(range(0,32))
ax.set_xticklabels(cars.car_names,rotation=60,fontsize='large')
ax.set_title('Miles Per Gallon Cars')
ax.set_xlabel("Car Names")
ax.set_ylabel("Mile Per/Gal")

```
