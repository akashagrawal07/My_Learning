# This script will create summary sheet for MFPAPI using report generated from clog performance tool 
1. When providing path make sure all files are extracted and kept in same folder from which script is running
2. Both Version should not be same (if comparing same version with different build try avoid using same name for version)

```python
import pandas as pd
import os
from datetime import datetime,timedelta
import xlsxwriter
from xlsxwriter import workbook

import matplotlib.pyplot as plt
from matplotlib import rcParams

def createIterationSheet(version,writer,final_List):
    print("Creating Iteration Sheet for version",version)
    count = 1
    Iteration = 1
    sheetName = f'{version}_Iteration'
    
    for df in final_List:
            
        df.to_excel(writer,sheetName,index=False,startrow=int(count+1))
        worksheet = writer.sheets[sheetName]
        worksheet.write(count, 0, f'Iteration-{Iteration}')    
        Iteration = Iteration + 1
        count = count + len(df) + 5

        setColumnFormat(sheetName,writer,0, 0, 60)
        setColumnFormat(sheetName,writer,1, 1, 15)
        setColumnFormat(sheetName,writer,2, 2, 25)

def createIterationSummarySheet(version,writer,final_List):
    print("Creating Iteration Summary Sheet for version ",version)
    totaltime = []

    count = 1
    sheetName = f'{version}_Iteration_Summary'    
    for df in final_List:
        df["Execution Time"].to_excel(writer,sheetName,index=False,startcol=count,header=[f'Iteration-{count}'])
        #Adding Data

        setColumnFormat(sheetName,writer,count, count)

        total = df['Execution Time'].sum()    
        totaltime.append(total)
        count = count + 1    

    return totaltime

def createFinalSummarySheet(writer,totaltime):
    print("Creating Final Summary...")
    dataFrame = pd.DataFrame()
    count = 1
    column = 4
#     print(totaltime)
    for version,time in totaltime.items():
#         print(version,time)
        dataFrame.to_excel(writer,"Final_Summary",index=False)
        worksheet = writer.sheets['Final_Summary']
        #Adding data 
        worksheet.write(1, column, f'{version}')
        
        count = 1
        for t in time:
            worksheet.write(1+count, 3, f'Iteration-{count}') 
            worksheet.write(1+count, column, t) 
            count = count + 1
            setColumnFormat('Final_Summary',writer,count, count)
            
        column = column + 1
        
    
def getColumnFormat(writer):
    workbook  = writer.book
    format1 = workbook.add_format({'num_format': '#,##0.00'})
    return format1

def setColumnFormat(sheetName,writerObj,first_col,last_col,width=13):
    worksheet = writerObj.sheets[sheetName]  
    worksheet.set_column(first_col,last_col, width, getColumnFormat(writerObj))

    
def createDataFrame(filepath):
        
    pf = pd.read_csv(filepath)

    my_columns = ['LogLine','Execution Time','Component']
    final_dataframe = pd.DataFrame(columns=my_columns)        

    t1 = (datetime.now().time())
    for i in pf.itertuples():#range(0,len(pf)):    

        if(i.ComponentName == 'MFPAPI' or i.ComponentName == 'AL - appmanagerlibrary'):

            #Changing Time in DataFrame in micorseconds
            date = pd.to_datetime(i.ExecutionTime[7:],infer_datetime_format=True)
            milisecond1 = (date.hour * 3600000 + date.minute * 60 * 1000 + date.second * 1000 + date.microsecond)/1000

#             #Creating DataFrame using Panda Series
            final_dataframe = final_dataframe.append(pd.Series([
                    i.LogLinePath,
                    milisecond1,
                    i.ComponentName
                ],index=my_columns),ignore_index=True)
    return final_dataframe

def plotIteration(filename,oldVersion,newVersion):
    
    xls  = pd.ExcelFile(filename)    
    df = pd.read_excel(xls,'Final_Summary')
    df.columns = ['1','2','3','Iteration',oldVersion,newVersion]
    
    df[oldVersion] = df[oldVersion][1:].astype(float)
    df[newVersion] = df[newVersion][1:].astype(float)
    
    df.drop(['1','2','3'],axis=1,inplace=True)
    
    
    fig = plt.figure()
    ax = fig.add_axes([.1,.1,1,1])
    ax.grid()
    
    #Setting Step size in x axis
    total = 0
    if(int(df[oldVersion].count()) > int(df[newVersion].count())):
        total = (int)(df[oldVersion].count())
    else:
        total = (int)(df[newVersion].count())
        
    a = range(1,total+2)
    ax.set_xticks(a)    
    
    plt.xlabel("Number of Iteration")
    plt.ylabel("Time in miliseconds")    
    plt.plot(df[oldVersion],label=oldVersion)
    plt.plot(df[newVersion],label=newVersion)
    plt.legend(loc='best')
    plt.show()

def main():

    old_version = input("Enter Old Version-")
    
    new_version = input("Enter New Version-")


    print("Enter folder path for Old Version and New Version")

    old_ver_path = input("Enter Old Version Folder Path-")
    while True:   
        if os.path.isdir(old_ver_path) == 0:
            print("Directory not exist -",old_ver_path)
            old_ver_path = input("Enter again Old Version Path-")
        else:
            break;

    new_ver_path = input("Enter New Version Folder Path-")
    while True:   
        if os.path.isdir(new_ver_path) == 0:
            print("Directory not exist -",new_ver_path)
            old_ver_path = input("Enter again New Version Path-")
        else:
            break;


    pathList = {}
    pathList[old_version] = old_ver_path
    pathList[new_version] = new_ver_path
    t1 = (datetime.now().time())
    print("Creating Summary sheet for both Version....")
    fileName = old_version + '_' + new_version + '_Summary_MFPAPI.xlsx'

    if os.path.isfile(fileName) == 1:
        os.remove(fileName)


    writer = pd.ExcelWriter(fileName,engine = 'xlsxwriter')
    totaltime = {}

    for version,path in pathList.items():
#         print(version,path)
        fname = []

        for root,d_names,f_names in os.walk(path):
            for f in f_names:
                fname.append(os.path.join(root, f))

        final_List = []

        for filepath in fname:    
            if(filepath.find('filtered.txt_final.csv') != -1):
                dataframe = createDataFrame(filepath)
                final_List.append(dataframe)
#         print(len(final_List)," DataFrame return from version ",version)
       
        createIterationSheet(version,writer,final_List)

        time = createIterationSummarySheet(version,writer,final_List)
        totaltime[version] = time
        print("\n")

    createFinalSummarySheet(writer,totaltime)
    writer.close()

    #plot the graph of Version
    plotIteration(fileName,old_version,new_version)

    print("Excel saved and closed bye...")
    t2 = (datetime.now().time())
    datetimeFormat = '%H:%M:%S.%f'
    t3 = (datetime.strptime(str(t2), datetimeFormat) - datetime.strptime(str(t1), datetimeFormat))
    print("Total Time Take to process-",t3)   
    

      
if __name__ == "__main__":
    print("This tool will create Summary Sheet for MFPAPI using report generated from clog tool")
    print("when providing path make sure all files are extracted\n\n")

    main()
```
